## CartPole 실행해 보기

```python
# 구현에 사용할 패키지 임포트
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import gym # 막대기 관련 물리법칙 구현. s_t -> s_t+1로 연결해 주는 장치. 카트 외 다양한 예시 지원. 
```

```python
# 애니메이션을 만드는 함수
from matplotlib import animation
from IPython.display import HTML


def display_frames_as_gif(frames):
    """
    Displays a list of frames as an HTML5 video.
    """
    plt.figure(figsize=(frames[0].shape[1]/72.0, frames[0].shape[0]/72.0),
               dpi=72)
    patch = plt.imshow(frames[0])
    plt.axis('off')

    def animate(i):
        patch.set_data(frames[i])

    anim = animation.FuncAnimation(plt.gcf(), animate, frames=len(frames),
                                   interval=50)

    anim.save('movie_cartpole.mp4')
    plt.close()

    video = open('movie_cartpole.mp4', 'rb').read()
    display(HTML(video_tag(video, 'mp4')))

# Function to generate HTML video tag
def video_tag(video, mimetype):
    """Generate a video tag for embedding video in HTML."""
    from base64 import b64encode
    base64_video = b64encode(video).decode("utf-8")
    video_tag = '<video controls alt="test" src="data:video/{0};base64,{1}">'.format(
        mimetype, base64_video)
    return video_tag
```

```python
# Generate frames
frames = []
env = gym.make('CartPole-v0')
observation = env.reset()

for step in range(0, 200):
    frames.append(env.render(mode='rgb_array'))
    action = np.random.choice(2) # policy : 좌우 랜덤하게 -> 설 리가 없음.
    observation, reward, done, info = env.step(action)
```

```python
display_frames_as_gif(frames)
```
