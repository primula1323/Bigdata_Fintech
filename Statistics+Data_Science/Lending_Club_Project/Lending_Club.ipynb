```python
def extract_NaN(str, df) :
    s1 = df[str]
    print(s1.isna().sum())
    return df[s1.isna()]
```

```python
def skim_var(str, df) :
    print(df[str])
    print(df[str].unique())
```

## 23. fico_range_high/24. fico_range_low

__fico_range_high__ : The upper boundary range the borrower’s FICO at loan origination belongs to.

__fico_range_low__ : The lower boundary range the borrower’s FICO at loan origination belongs to.

* fico_range_high는 대출 개시 시점 차주의 'FICO'의 최댓값을, fico_range_low는 대출 개시 시점 차주의 'FICO'의 최솟값을 의미함.

* FICO : FICO사(Fair Issac Corporation)가 결제 기록, 현재 채무, 채무 종류, 채무 기간, 신규 채무 계좌 등을 활용해 예측한 차주의 신용도. '신용점수'로 생각하면 됨.(자세한 정보는 https://www.investopedia.com/terms/f/ficoscore.asp 참고)

```python
skim_var('fico_range_high', df)
```

```python
skim_var('fico_range_low', df)
```

```python
extract_NaN('fico_range_high', df)
```

```python
extract_NaN('fico_range_low', df)
```

## 28. home_ownership

```python
print(sum(df['home_ownership'] == 'NONE'))
print(sum(df['home_ownership'] == 'ANY'))
print(sum(df['home_ownership'] == 'OTHER'))
extract_NaN('home_ownership', df)
```

## 31. initial_list_status(A/B test dummy)

__initial_list_status__ : The initial listing status of the loan. * Possible values are – W, F


```python
skim_var('initial_list_status', df)
```

W와 F는 각각 무슨 의미인가? 

* W : Whole
* F : Fractional

이는 https://www.fintechnexus.com/lending-club-whole-loan-program-one-year-later/를 참고하여 결론을 내렸음. 즉, 해당 변수는 lending club 사의 A/B test를 통해 할당된 변수이고, orthogoanal to other variables여야 함.

보다 구체적으로, 해당 변수는 lending club 사에서 시행했던 '전체 대출을 한 번의 모금으로 해결'과 '여러 사람의 모금으로 해결'의 A/B test를 나타내는 변수임. 이는 각 대출별로 random assignment이므로 사용 불가.

```python
extract_NaN('initial_list_status', df)
```
