# Practice 2.1 Quotient and remainder

## Create a function for calculating the quotient and remainder

```python
def my_devide(num1: int, num2: int)->int:
    quotient = num1 // num2
    remainder = num1 % num2
    return quotient, remainder
```

```python
def my_devide(num1: int, num2: int)->int:
    quotient, remainder = divmod(num1, num2)
    return quotient, remainder
```

```python
def my_devide(num1: int, num2: int)->int:
    quotient = int(num1/num2)
    remainder = num1 - num2*quotient
    return quotient, remainder
```

# Practice 2.2 Number of diagonals in a polygon

## Create a function for calculating the Number of diagonals in a polygon

```python
def my_diagonal(num1: int)->int:
    diagonal = int(num1*(num1-3)/2)
    return diagonal
```

```python
my_diagonal(5)
```

# Practice 2.3 Rounding

## Create a function for rounding to an integer

```python
def my_rounding(num1: float)->int:
    if (num1>0):
        rounding = round(num1-0.5)
    rounding = round(num1+0.5)
    return rounding
```

```python
def my_rounding(num1: float)->int:
    if (num1>0):
        rounding = int(num1-0.5)
    rounding = int(num1+0.5)
    return rounding
```

```python
my_rounding(4.5)
```

```python
for i in range(-4,5):
    print(i-0.5, my_rounding(i-0.5))
```

# Pratice 2.1 Name Separator

### Taking a name as input, write a function that separates last name and first name.

```python
def name_separator(name):

    #split first name and last name
    separated = name.split()
    first_name = separated[0]
    last_name = separated[-1]

    print(f"first name is {first_name}")
    print(f"last name is {last_name}")

my_name = input("Write your name:(eg.Sun-sin Yi) ")
name_separator(my_name)
```

# Pratice 2.2. Age and Gender Calculator

### Taking the first seven digits of the residential number as input values, write a function that outputs the age and gender.

**Guidelines**
1) Distinguish gender (hint: 7th digit: 1 or 3- male, 2 or 4-female) \
2) Calculate the age based on today's date (hint:compare birthday and today's date)

```python
from datetime import date
today = date.today()
today = today.strftime("%Y%m%d")


def q3(residential_number):

    #initialization (optional)
    age=0
    gender=''

    #split numbers for its purpose (age, gender)
    num4age = residential_number[:6] #str
    num4gender=residential_number[-1] #str

    #step1) gender
    if num4gender in('1','3'):
        gender='male'
    else:
        gender='female'

    #step2)
    #changing format to calculate age
    if num4gender in('3','4'):
        birth ='20'+ num4age
    else :
        birth ='19'+ num4age


    ##cases that birthday has not passed (-1)
    if int(birth[4:]) > int(today[4:]):
        age=int(today[:4])-int(birth[:4])-1
    ##cases that birthday has passed
    else:
        age=int(today[:4])-int(birth[:4])

    print(f"Age: {age}")
    print(f"Gender: {gender}")


my_number = input("Write your residential number:(eg.220908-2)  ")
q3(my_number)
```
