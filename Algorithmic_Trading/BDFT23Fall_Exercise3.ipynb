### Exponentially-weighted moving average (p.205)

- Whereas in the simple moving average the past observations are weighted equally, exponential functions are used to assign exponentially decreasing weights over time. 

- For more information about EWMA, see the following link : https://en.wikipedia.org/wiki/Exponential_smoothing

- $s_0 = x_0$,
- $s_t = \alpha x_t + (1-\alpha)s_{t-1}$

```python
# Year 2005~2009
aapl_px = close_px.AAPL['2005':'2009']
# 60-day moving average of AAPL stock price
# min_periods=50 : at least 50 non-NA values within the window
ma60 = aapl_px.rolling(60, min_periods=50).mean()

# 60-day moving average of AAPL stock price with exponential weighting
ewma60 = aapl_px.ewm(span=60).mean() # alpha = 2 / (span +1) for span >= 1

# Plot the results
fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True, sharey=True, figsize=(12,7))
aapl_px.plot(style='k-', ax=axes[0], label='Series')
ma60.plot(style='r:', ax=axes[0], label='MA60')
axes[0].legend(loc='upper left')
axes[0].set_title('Simple MA')
aapl_px.plot(style='g-', ax=axes[1], label='Series')
ewma60.plot(style='b--', ax=axes[1], label='EWMA60')
axes[1].set_title('Exponentially-weighted MA')
axes[1].legend(loc='upper left')
plt.show()
```
